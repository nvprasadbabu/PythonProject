```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
z = np.linspace(-1,1,20)
z
```

## Identity Function:

```python
# This function lets the activation value go through
def f(x):
    return x
```

```python
output_if = [f(x) for x in z]
```

```python
plt.plot(z, output_if, label="Identity Function")
plt.xlabel("Z")
plt.ylabel("f(z)")
plt.legend(loc="best")    
```

## Step Function:

```python
# This function activates the neuron if the activation if the activation
# is above a certain value that is called threshold activity function
def sf(x):
    if x >= 0:
        output = 1
    else:
        output = 0
    return output    
```

```python
output_sf = [sf(x) for x in z]
```

```python
plt.plot(z, output_sf, label="Step Function")
plt.xlabel("Z")
plt.ylabel("f(z)")
plt.legend(loc="best")
```

## Sigmoid Function:

```python
# logistic function
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))
```

```python
X_sig = np.linspace(-10,10,100)
```

```python
output_sigmoid = [sigmoid(x) for x in X_sig]
```

```python
plt.plot(X_sig, output_sigmoid, label="Sigmoid Function")
plt.xlabel("Z")
plt.ylabel("f(z)")
plt.legend(loc="best")
```

## Hyperbolic Tangent:

```python
output_ht = [np.tanh(x) for x in z]
```

```python
plt.plot(z, output_ht, label="Hyberbolic Tangent Function")
plt.xlabel("Z")
plt.ylabel("f(z)")
plt.legend(loc="best")
```

## ReLU:

```python
#Rectified Linear Unit
def relu(x):
    if x >= 0:
        output = x
    else:
        output = 0
    return output  
```

```python
output_relu = [relu(x) for x in z]
```

```python
plt.plot(z, output_relu, label="ReLU Function")
plt.xlabel("Z")
plt.ylabel("f(z)")
plt.legend(loc="best")
```

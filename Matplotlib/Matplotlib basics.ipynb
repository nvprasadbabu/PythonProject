## Basic Plot:

```python
X = np.linspace(-np.pi, np.pi, 300, endpoint=True)
sin,cos = np.sin(X), np.cos(X)
plt.figure(figsize=(9,6))
plt.plot(X,sin,color="red", linewidth=2, linestyle="--")
plt.plot(X,cos,color="green", linewidth=1, linestyle="-")

plt.show()
```

### Setting Limits:

```python
#setting limits:
plt.xlim(X.min()*1.2, X.max()*1.2)
plt.ylim(cos.min()*1.1, cos.max()*1.1)
```

### Setting Ticks:

```python
#setting ticks:
plt.xticks( [-np.pi, -np.pi/2, -np.pi/4, 0, np.pi/4, np.pi/2, np.pi])
plt.yticks([-1, -0.5, 0, 0.5, +1])
```

### Setting tick labels:

```python
#setting tick labels
plt.xticks([-np.pi, -np.pi/2, -np.pi/4, 0, np.pi/4, np.pi/2, np.pi],
       [r'$-\pi$', r'$-\pi/2$', r'$-\pi/4$',r'$0$', r'$-\pi/4$',r'$+\pi/2$', r'$+\pi$'])
```

### Adding a Legend:

```python
#adding a legend
plt.plot(X, cos, color="blue", linewidth=2, linestyle="--", label="Cosine(X)")
plt.plot(X, sin, color="red",  linewidth=1.5, linestyle="-", label="Sine(X)")

plt.legend(loc='best',frameon=True)
```

### Adding Annotations:

```python
#adding annotations
plt.plot(X, cos, color="blue", linewidth=2, linestyle="--", label="Cosine(X)")
plt.annotate('local max', xy=(0, 1.0), xytext=(1.0, 0.8),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )
plt.plot(X, sin, color="red",  linewidth=1.5, linestyle="-", label="Sine(X)")

plt.annotate('local min', xy=(-1.6, -1.0), xytext=(0, -0.8),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )
plt.legend(loc='best',frameon=True)
```

### Drawing Accumulative Areas:

```python
#drawing accumulative areas
plt.fill_between(X, 0, cos, cos > 0, color='blue', alpha=.25)
plt.fill_between(X, -1, sin, sin < 0, color='red',  alpha=.25)
```

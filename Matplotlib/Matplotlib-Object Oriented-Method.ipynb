## Object Oriented Method:

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.arange(0, 10, 0.2)
y = np.sin(x)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y)
ax.set_title("Graph")
ax.set_xlabel("X")
ax.set_ylabel("Sin(X)")
plt.show()
```

```python
x = np.linspace(-10,10,20)
fig, axes = plt.subplots(nrows=1, ncols=3,figsize=(10, 5))

axes[0].plot(x, x**2 - x, x, x**3)
axes[0].set_title("Polynomial Functions")

axes[1].plot(x, x**2 + 1.5, x, x**4+ x**3-x**2+1)
axes[1].set_title("Polynomial Functions")

axes[2].plot(x, x*2+0.5, x, x**3)
axes[2].set_ylim([0, 30])
axes[2].set_xlim([0, 10])
axes[2].set_title("Polynomial Functions")
```

### Text Annotation:

```python
#text annotation
axes[2].text(3, 5, r"$y=2x+0.5$", fontsize=15, color="blue")
axes[2].text(2.5, 10, r"$y=x^2$", fontsize=15, color="green")
fig
```

## Creating a helper function to plot a graph

```python
def plotter(ax, data1, data2, param_dict):
    """
    ax : Axes
        The axes to draw to

    data1 : array that contains the x data

    data2 : array that contains the y data

    param_dict : dict
       Dictionary of kwargs to pass to ax.plot

    Returns
    -------
    out : list
        list of artists added
    """
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python
fig, ax = plt.subplots(1, 1)
x = np.arange(0, 10, 0.2)
y = np.sin(x)
plotter(ax, x, y, {'marker':'x'})
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
plotter(ax1, x, y, {'marker':'x'})
plotter(ax2, x, y, {'marker':'o'})
```

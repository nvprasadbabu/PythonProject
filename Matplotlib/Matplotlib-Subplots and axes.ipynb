## SubPlots and Axes:

### SubPlots: You need to specify the number of rows and columns and the number of the plot.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x1 = np.linspace(0,5,20)
y1 = np.exp(x1) - x1
y2 = x1 - np.sqrt(x1)
plt.subplot(1,2,1)
plt.plot(x1, y1, 'r--')
plt.subplot(1,2,2)
plt.plot(x1, y2, 'g*-')
plt.show()
```

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
```

```python
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
plt.subplot(2, 1, 1)

plt.plot(x1, y1, 'o-')
plt.title('2 subplots')
plt.ylabel('oscillation')

plt.subplot(2, 1, 2, facecolor='y') # facecolor : background color.
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('')

plt.show()
```

```python
## Adding Annotations:
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)
#Annotation
plt.annotate('local max', xy=(1, 0.4), xytext=(1.5, 0.8),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )
plt.plot(x1, y1, 'o-')
plt.title('2 subplots')
plt.ylabel('oscillation')

plt.subplot(2, 1, 2, facecolor='y') # facecolor : background color.
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('')

plt.show()
```

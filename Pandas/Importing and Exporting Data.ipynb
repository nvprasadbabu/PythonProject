### To write a data to the .csv file, use to_csv function

```python
d = {'c1': pd.Series(['a', 'b', 'c', 'd']),
    'c2': pd.Series([1,2,3,4])}
df = pd.DataFrame(d)
df.to_csv('../../data/samplecsv.csv',index=False)
```

```python
df=pd.read_csv('../../data/samplecsv.csv')
df
```

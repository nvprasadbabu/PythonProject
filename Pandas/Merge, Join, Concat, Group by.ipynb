<center><u><H1>Pandas: Merge, Join, Concat and Group by</H1></u></center>

```python
import pandas as pd
```

```python
data1 = {
        'id': ['1', '2', '3', '4', '5'],
        'first_name': ['Sam', 'Betty', 'Adam', 'Alice', 'Bill'], 
        'last_name': ['Hills', 'Li', 'Roberts', 'Chan', 'Howard']}
df1 = pd.DataFrame(data1, columns = ['id', 'first_name', 'last_name'])
df1
```

```python
data2 = {
        'id': ['4', '5', '6', '7', '8'],
        'first_name': ['Britany', 'Cecil', 'Arthur', 'Carl', 'George'], 
        'last_name': ['Black', 'Berninson', 'Baldwin', 'Sagan', 'Smith']}
df2 = pd.DataFrame(data2, columns = ['id', 'first_name', 'last_name'])
df2
```

```python
data3 = {
        'id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
df3 = pd.DataFrame(data3, columns = ['id','test_id'])
df3
```

## JOIN METHODS: (MERGE AND CONCAT)

### Merge method--->SQL Join Name---->	Description
##### left---------------------->LEFT OUTER JOIN--------->Use keys from left frame only
##### right--------------------> RIGHT OUTER JOIN------->Use keys from right frame only
##### outer------------------->	FULL OUTER JOIN--------->Use union of keys from both frames
##### inner------------------->	INNER JOIN------------------->Use intersection of keys from both frames

```python
# Join two dataframes along rows
df_rows = pd.concat([df1, df2])
df_rows
```

```python
# Join two dataframes along columns
df_col = pd.concat([df1, df2], axis=1)
df_col
```

```python
# append function
new_df1 = df1.append(df2)
new_df1
```

### Inserting a key argument

```python
dflist = [df1,df2]
ckey = pd.concat(dflist, keys=['a', 'b'])
ckey
```

## MERGE:

### pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None,
### left_index=False, right_index=False, sort=True,  suffixes=('_x', '_y'), copy=True, indicator=False) 

```python
df_m = pd.merge(df1, df3, on='id')
df_m
```

```python
df_m1 = pd.merge(df3, df1, on='id')
df_m1
```

```python
#Merge two dataframes with both the left and right dataframes using the key
df_m2 = pd.merge(df_rows, df3, left_on='id', right_on='id')
df_m2
```

```python
#Merge with outer join
df_m3 = pd.merge(df1, df2, on='id', how='outer')
df_m3
```

```python
#Merge with inner join
df_m4 = pd.merge(df1, df2, on='id', how='inner')
df_m4
```

```python
#Merge with right join
df_m5 = pd.merge(df1, df2, on='id', how='right')
df_m5
```

```python
#Merge with left join
df_m6 = pd.merge(df1, df2, on='id', how='left')
df_m6
```

```python
#Merge while adding a suffix to duplicate column names
df_m7 = pd.merge(df1, df2, on='id', how='left', suffixes=('_left', '_right'))
df_m7
```

## JOIN:

```python
dfj = df1.join(df2, lsuffix='_df1', rsuffix='_df2')
dfj
```

```python
#If we want to join using the key columns, we need to set key to be the index in both df.
#The joined DataFrame will have key as its index.
dfj1 = df1.set_index('id').join(df2.set_index('id'),lsuffix='_df1', rsuffix='_df2')
dfj1
```

```python
dfj2 = df1.join(df2.set_index('id'), on='id',lsuffix='_df1', rsuffix='_df2')
dfj2
```

## GROUP BY:

```python
df = pd.DataFrame({'FactorA': [.6, .9, .3, 1.5],
                    'FactorB': [41, 43, 4, 15],
                    'FactorC': [1, 3, 2, 6],
                    'Class': ['0', '1', '1', '0']},index=[0, 1, 2, 3])
df
```

```python
df.groupby('Class').sum()
```

```python
df.groupby(['Class']).mean()
```

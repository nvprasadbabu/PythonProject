<center><u><H1>Seaborn-Color palette</H1></u></center>

```python
import seaborn as sns
%matplotlib inline
```

## Qualitative color palettes

```python
palette = sns.color_palette()
sns.palplot(palette)
```

```python
#deep, muted, pastel, bright, dark, and colorblind
sns.palplot(sns.color_palette('muted'))
```

```python
sns.palplot(sns.color_palette('pastel'))
```

```python
sns.palplot(sns.color_palette('dark'))
```

## Circular color system

```python
sns.palplot(sns.color_palette("hls", 8))
```

```python
sns.palplot(sns.hls_palette(8, l=.5, s=.9)) # l for lightness and s for saturation
```

```python
sns.palplot(sns.color_palette("husl", 8))
```

```python
sns.palplot(sns.husl_palette(8, l=.5, s=.9))
```

## Categorical Color Brewer palettes

```python
sns.palplot(sns.color_palette("Paired",8))
```

```python
sns.palplot(sns.color_palette("Set2", 10))
```

```python
flatui = ["#9b59b6", "#3498db", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"]
sns.palplot(sns.color_palette(flatui))
```

## Sequential Color Palettes

```python
sns.palplot(sns.color_palette("BuGn_r"))
```

```python
sns.palplot(sns.color_palette("GnBu_d"))
```

## Diverging Color Palettes

```python
sns.palplot(sns.color_palette("BrBG", 7))
```

```python
sns.palplot(sns.color_palette("RdBu_r", 7))
```

```python
sns.palplot(sns.color_palette("coolwarm", 7))
```

## Setting the default color pallete

```python
sns.set_palette("husl")
sns.palplot(sns.color_palette())
```

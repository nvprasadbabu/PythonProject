## How To Use Seaborn’s Colors As A colormap in Matplotlib?

```python
from matplotlib.colors import ListedColormap
import numpy as np

# Number of samples
N = 100

# Construct the colormap
palette = sns.color_palette("muted", n_colors=5)
cmap = ListedColormap(sns.color_palette(palette).as_hex())# hexadecimal casting

# Initialize the data
x1 = np.random.randn(N)
x2 = np.random.randn(N)
# Assume that there are 5 possible labels
colors = np.random.randint(0,5,N)

# Create a scatter plot
plt.scatter(x1, x2, c=colors, cmap=cmap)

# Add a color bar
plt.colorbar()

# Show the plot
plt.show()
```

## How To Set The Figure Size in Seaborn

```python
#For axes level functions, you can make use of the plt.subplots() function to which you pass the figsize argument
#Let's see an example with categorical plots.
#For Figure-level functions, you rely on two parameters to set the Figure size, namely, size and aspect
```

```python
#axes level function:
fig, ax = plt.subplots(figsize=(10,4))
sns.stripplot(x="species", y="petal_length", data=iris, ax=ax)
```

```python
g = sns.catplot(x="species", y="petal_length", data=iris, kind="bar", height=4, aspect=2, hue="species", palette="muted", legend=False)
```

## How To Rotate Label Text in Seaborn

```python
g = sns.catplot(x="species", y="petal_length", data=iris, kind="bar", height=4, aspect=2, hue="species", palette="muted", legend=False)
g.set_xticklabels(rotation=90)
```

## Setting xlim or ylim in Seaborn

```python
g = sns.catplot(x="species", y="petal_length", data=iris, kind="bar", height=4, aspect=2, hue="species", palette="muted", legend=False)
g.set_xticklabels(rotation=90)
g.set(ylim=(0, 10))
# alternative you can use : ax.set_ylim(0,10)
```

## Figure-level:

```python
# you can access the Axes object
axes = g.ax
g = sns.catplot(x="species", y="petal_length", data=iris, kind="bar", height=4, aspect=2, hue="species", palette="muted", legend=False)
g.set_xticklabels(rotation=90)

#accesing to the 'Axes' properties
axes.set_ylim(0,5.5)
```

## Adding a Title

```python
g = sns.catplot(x="species", y="petal_length", data=iris, kind="bar", height=4, aspect=2, hue="species", palette="muted", legend=False)
g.set_xticklabels(rotation=90)
fig = g.fig
# Add a title to the Figure
fig.suptitle('Petal Length for species', fontsize=12)
```

```python
b = sns.boxplot(x="species", y="petal_length", data=iris, hue="species", palette="muted")
# add a title
b.set_title('Bar graph for species')
```

## How To Scale Seaborn Plots

```python
# Reset default params
sns.set()
# Available contexts are: paper, notebook, talk and poster
# Set context to "notebook"
sns.set_context("notebook")
b = sns.boxplot(x="species", y="petal_length", data=iris, hue="species", palette="muted")
```

```python
#You can also pass more arguments to set_context() 
sns.set_context("paper", font_scale=2, rc={"font.size":6,"axes.labelsize":6})
sns.boxplot(x="species", y="petal_length", data=iris, hue="species", palette="muted")
```

```python
# Reset default params
sns.set(rc={"font.size":6,"axes.labelsize":6})
sns.boxplot(x="species", y="petal_length", data=iris, hue="species", palette="muted")
```

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Plotting univariate distributions:

```python
# displot function
f = np.random.normal(size=150)
sns.displot(f)
```

### Histograms:

#### A histogram represents the distribution of data by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin. we also can plot a rug plot, which draws a small vertical tick at each observation.

```python
sns.histplot(f, kde=False)
```

### Kernel Density Estimation:

```python
sns.displot(f,  rug=True)
```

```python
sns.kdeplot(f, fill=True)
```

### Hexbin plots:
it shows the counts of observations that fall within hexagonal bins

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
setosa = iris.loc[iris.species == "setosa"]
virginica = iris.loc[iris.species == "virginica"]
```

```python
sns.jointplot(x=setosa.sepal_width, y=setosa.sepal_length,
                  cmap="Reds",color='r', kind="hex")
#kind : { “scatter” | “reg” | “resid” | “kde” | “hex” }
```

### Kernel density estimation:

#### you can also plot two-dimensional kernel density plot

```python
sns.kdeplot(x=setosa.sepal_width, y=setosa.sepal_length,
                  cmap="Blues", fill=True, cbar=True)
```

```python
sns.kdeplot(x=setosa.sepal_width, y=setosa.sepal_length,
                  cmap="Reds", fill=True, thresh=0.05)
sns.kdeplot(x=virginica.sepal_width, y=virginica.sepal_length,
                  cmap="Blues", fill=True, thresh=0.05)
```

```python
sns.jointplot(x="sepal_width", y="petal_length", data=iris,kind="kde", color="g")
```

### Visualizing pairwise relationships in a dataset:

```python
sns.pairplot(iris,palette='husl',hue='species',markers=["o", "s", "D"])
```

### Linear Regression:

```python
sns.jointplot(x=setosa.sepal_width, y=setosa.sepal_length,color='r', kind="reg")
```

```python
sns.pairplot(iris, kind="reg", height=2, aspect=2)
```

```python
sns.pairplot(iris, kind="reg", hue="species", palette="husl", height=2.5, aspect=2)
```

## Heat Map:

### Plotting a correlation matrix:

```python
sns.heatmap(iris.corr(numeric_only=True), annot=True, linewidths=.5, cmap="YlGnBu")
# if you dont need the colobar:cbar=False
```

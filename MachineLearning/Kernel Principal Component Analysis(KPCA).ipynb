<center><u><H1>Kernel Principal Component Analysis(KPCA)</H1></u></center>

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_circles
from sklearn.decomposition import KernelPCA
from sklearn.decomposition import PCA
%matplotlib inline
```

```python
X, y = make_circles(n_samples=1000, noise=0.10, factor=0.2)
```

```python
X.data.shape
```

```python
y.data.shape
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X[y==0, 0], X[y==0, 1], color='green', marker='v')
plt.scatter(X[y==1, 0], X[y==1, 1], color='blue', marker='o')
plt.title('Concentric circles')
plt.ylabel('X1')
plt.xlabel('X0')
plt.show()
```

## Principal Component Analysis (PCA)

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_pca[y==0, 0], np.zeros((500,1))+0.1, color='red', alpha=0.5)
plt.scatter(X_pca[y==1, 0], np.zeros((500,1))-0.1, color='blue', alpha=0.5)
plt.ylim([-5,5])
plt.title('First principal component after Linear PCA')
plt.xlabel('PC1')
plt.show()
```

## KPCA

```python
kpca = KernelPCA(kernel='rbf', fit_inverse_transform=True, gamma=10)
X_kpca = kpca.fit_transform(X)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_kpca[y==0, 0], X_kpca[y==0, 1], color= 'red', marker='v')
plt.scatter(X_kpca[y==1, 0], X_kpca[y==1, 1], color= 'blue', marker='o')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_kpca[y==0, 0], np.zeros((500,1))+0.02, color='red')
plt.scatter(X_kpca[y==1, 0], np.zeros((500,1))-0.02, color='blue')
plt.ylim([-5,5])
plt.title('First principal component after Kernel PCA')
plt.xlabel('PC1')

plt.show()
```

```python
X_back = kpca.inverse_transform(X_kpca)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_back[y==0, 0], X_back[y==0, 1], color= 'purple', marker='x')
plt.scatter(X_back[y==1, 0], X_back[y==1, 1], color= 'yellow', marker='v')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

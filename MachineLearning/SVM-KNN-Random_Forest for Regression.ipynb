# <center><u>SVM, KNN and Random Forest for Regression</u></center>

```python
#from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
X = data
y = target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2018)
```

```python
svr = SVR(kernel='linear')
```

```python
svr_rbf = SVR(kernel='rbf', C=1000, gamma=0.001)
```

```python
knn_r = KNeighborsRegressor(n_neighbors=1)
```

```python
rf_r = RandomForestRegressor(n_estimators=100, random_state=2018)
```

## Support Vector Machines for Regression:

```python
svr.fit(X_train,y_train)
svr_pred = svr.predict(X_test)
```

```python
print("R2 = {:.2}".format(r2_score(y_test, svr_pred)))
```

```python
svr_rbf.fit(X_train,y_train)
svr_rbf_pred = svr_rbf.predict(X_test)
```

```python
print("R2 = {:.2}".format(r2_score(y_test, svr_rbf_pred)))
```

## K-Nearest Neighbors for Regression:

```python
knn_r.fit(X_train, y_train)
knn_pred = knn_r.predict(X_test)
```

```python
print("R2 = {:.2}".format(r2_score(y_test, knn_pred)))
```

## Random Forest for Regression:

```python
rf_r.fit(X_train, y_train)
rf_pred = rf_r.predict(X_test)
```

```python
print("R2 = {:.2}".format(r2_score(y_test, rf_pred)))
```
